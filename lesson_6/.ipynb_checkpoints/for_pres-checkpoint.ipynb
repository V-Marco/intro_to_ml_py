```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
data = load_boston(return_X_y=True)
boston = load_boston()
```

```python
X = pd.DataFrame(data[0])
Y = data[1]
```

```python
X.columns = boston['feature_names']
```

```python
plt.scatter(X['RM'], Y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(np.array(X['RM']).reshape(-1, 1), Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
Y_new = model.predict(np.array(X['RM']).reshape(-1, 1))
```

```python
import seaborn as sns
sns.set_style('darkgrid')
plt.scatter(X['RM'], Y, alpha=0.7)
plt.plot(X['RM'], Y_new, c='orange')
plt.title('Зависимость цены на жильё от среднего числа комнат в квартире')
plt.xlabel('Среднее число комнат в квартире')
plt.ylabel('Цена на жильё')
```
